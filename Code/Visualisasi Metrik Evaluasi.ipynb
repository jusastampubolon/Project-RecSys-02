Import Libraries

```python
# Import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

Data Evaluasi untuk Masing-Masing Model

```python
# Data evaluasi untuk masing-masing model (CBF, User-Based CF, TF-IDF)

# Hasil evaluasi model Collaborative-Based Filtering (CBF)
cbf_rmse = [0.13]
cbf_mae = [0.10]
cbf_accuracy = [0.80]
cbf_precision = [0.75]
cbf_recall = [1.00]
cbf_map = [0.60]

# Hasil evaluasi model User-Based Collaborative Filtering (User-Based CF)
user_based_cf_rmse = [0.17]
user_based_cf_mae = [0.14]
user_based_cf_accuracy = [0.60]
user_based_cf_precision = [0.67]
user_based_cf_recall = [0.67]
user_based_cf_map = [0.64]

# Hasil evaluasi model Term Frequency-Inverse Document Frequency (TF-IDF)
tfidf_rmse = [0.19]
tfidf_mae = [0.18]
tfidf_accuracy = [0.60]
tfidf_precision = [0.67]
tfidf_recall = [0.67]
tfidf_map = [0.60]
```

- Data evaluasi yang dimasukkan adalah hasil dari metrik-metrik seperti RMSE, MAE, Accuracy, Precision, Recall, dan MAP (Mean Average Precision).
- Setiap model memiliki nilai evaluasi yang berbeda, dengan CBF sebagai model yang diusulkan.

Membuat DataFrame untuk Metrik Evaluasi

```python
# Create DataFrame for evaluation metrics for each model

metrics = pd.DataFrame({
    'Metric': ['RMSE', 'MAE', 'Accuracy', 'Precision', 'Recall', 'MAP'],
    'CBF': [np.mean(cbf_rmse), np.mean(cbf_mae), np.mean(cbf_accuracy), np.mean(cbf_precision), np.mean(cbf_recall), np.mean(cbf_map)],
    'User-Based CF': [np.mean(user_based_cf_rmse), np.mean(user_based_cf_mae), np.mean(user_based_cf_accuracy), np.mean(user_based_cf_precision), np.mean(user_based_cf_recall), np.mean(user_based_cf_map)],
    'TF-IDF': [np.mean(tfidf_rmse), np.mean(tfidf_mae), np.mean(tfidf_accuracy), np.mean(tfidf_precision), np.mean(tfidf_recall), np.mean(tfidf_map)]
})

metrics
```

membuat DataFrame menggunakan pandas untuk menyimpan hasil evaluasi dari ketiga model. Metrik yang digunakan adalah RMSE, MAE, Accuracy, Precision, Recall, dan MAP.

- CBF dan User-Based CF tidak memiliki metrik MAP, sehingga nilai NaN digunakan untuk kedua model tersebut.
- TF-IDF memiliki nilai MAP yang tersedia, sehingga metrik tersebut dihitung dan dimasukkan dalam DataFrame.

Visualisasi RMSE

```python
# Plot RMSE for each model
fig, ax = plt.subplots(figsize=(6, 4))
sns.barplot(x=['CBF', 'User-Based CF', 'TF-IDF'], y=[np.mean(cbf_rmse), np.mean(user_based_cf_rmse), np.mean(tfidf_rmse)], ax=ax, palette='Blues')
ax.set_title('RMSE')
ax.set_ylabel('RMSE')
plt.show()
```

visualisasi metrik RMSE (Root Mean Squared Error) untuk masing-masing model menggunakan grafik batang. RMSE adalah metrik yang digunakan untuk mengukur seberapa baik model dalam memprediksi nilai yang benar.

CBF menunjukkan nilai RMSE terbaik dibandingkan dengan User-Based CF dan TF-IDF.

Visualisasi MAE

```python
# Plot MAE for each model
fig, ax = plt.subplots(figsize=(6, 4))
sns.barplot(x=['CBF', 'User-Based CF', 'TF-IDF'], y=[np.mean(cbf_mae), np.mean(user_based_cf_mae), np.mean(tfidf_mae)], ax=ax, palette='Greens')
ax.set_title('MAE')
ax.set_ylabel('MAE')
plt.show()
```

visualisasikan metrik MAE (Mean Absolute Error) untuk ketiga model. MAE menunjukkan rata-rata kesalahan absolut yang dibuat oleh model. Semakin rendah nilai MAE, semakin baik model tersebut dalam memprediksi hasil yang benar.

Dalam hal ini, CBF menunjukkan performa terbaik dengan MAE yang paling rendah.

Visualisasi Accuracy

```python
# Plot Accuracy for each model
fig, ax = plt.subplots(figsize=(6, 4))
sns.barplot(x=['CBF', 'User-Based CF', 'TF-IDF'], y=[np.mean(cbf_accuracy), np.mean(user_based_cf_accuracy), np.mean(tfidf_accuracy)], ax=ax, palette='Oranges')
ax.set_title('Accuracy')
ax.set_ylabel('Accuracy')
plt.show()
```

visualisasikan metrik Accuracy untuk masing-masing model. Accuracy mengukur persentase prediksi yang benar oleh model.

CBF memiliki accuracy yang lebih tinggi dibandingkan dengan User-Based CF dan TF-IDF.

Visualisasi Precision

```python
# Plot Precision for each model
fig, ax = plt.subplots(figsize=(6, 4))
sns.barplot(x=['CBF', 'User-Based CF', 'TF-IDF'], y=[np.mean(cbf_precision), np.mean(user_based_cf_precision), np.mean(tfidf_precision)], ax=ax, palette='Purples')
ax.set_title('Precision')
ax.set_ylabel('Precision')
plt.show()
```

visualisasikan metrik Precision untuk masing-masing model. Precision mengukur proporsi prediksi yang benar di antara semua prediksi positif yang dibuat oleh model.

CBF memiliki precision yang lebih tinggi dibandingkan dengan kedua model lainnya.

Visualisasi Recall

```python
# Plot Recall for each model
fig, ax = plt.subplots(figsize=(6, 4))
sns.barplot(x=['CBF', 'User-Based CF', 'TF-IDF'], y=[np.mean(cbf_recall), np.mean(user_based_cf_recall), np.mean(tfidf_recall)], ax=ax, palette='Reds')
ax.set_title('Recall')
ax.set_ylabel('Recall')
plt.show()
```

visualisasikan metrik Recall untuk masing-masing model. Recall mengukur proporsi hasil positif yang benar-benar berhasil diprediksi oleh model.

CBF memiliki nilai recall yang lebih tinggi, yang menunjukkan kemampuannya untuk mengenali sebagian besar hasil positif.

Visualisasi MAP (Mean Average Precision)

```python
# Plot MAP (Mean Average Precision) for each model
fig, ax = plt.subplots(figsize=(6, 4))
sns.barplot(x=['CBF', 'User-Based CF', 'TF-IDF'], y=[np.mean(cbf_map), np.mean(user_based_cf_map), np.mean(tfidf_map)], ax=ax, palette='cividis')
ax.set_title('MAP')
ax.set_ylabel('MAP')
plt.show()
```

visualisasikan metrik MAP (Mean Average Precision), yang hanya tersedia untuk model TF-IDF. MAP mengukur kualitas dari ranking hasil yang diberikan oleh model.

Karena CBF dan User-Based CF tidak memiliki metrik MAP, nilai NaN ditampilkan untuk kedua model tersebut, sementara TF-IDF menunjukkan nilai MAP yang dihitung.
